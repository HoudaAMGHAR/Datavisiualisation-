```python

# Filtrer les données pour ne conserver que les budgets supérieurs ou égaux à 4 millions
df_filtered = data[data['budget'] >= 4000000]

# Vérifier la valeur minimale du budget dans les données filtrées
min_budget = min(df_filtered['budget'])

# Tracer la boîte à moustaches horizontale
plt.figure(figsize=(10, 8))
plt.boxplot(df_filtered['budget'], vert=False)
plt.xlabel('Budget')
plt.title('Répartition du Budget (Budget >= 4 millions)'.format(min_budget))  # Titre du graphique avec le minimum du budget
plt.show()
```

<font color="orange"><b><h1> 5.Evolution des revenus medians des films par année de sortie (2000 à 2022) </h1></b></font>

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats


data['release_date'] = pd.to_datetime(data['release_date'])
filtered_data = data[(data['release_date'].dt.year >= 2000) & (data['release_date'].dt.year <= 2022)]
z_scores = np.abs(stats.zscore(filtered_data['revenue']))

# Définissons un seuil z-score au-delà duquel une valeur est considérée comme une valeur aberrante
threshold = 3

# Supprime les lignes contenant des valeurs aberrantes
filtered_data = filtered_data[(z_scores < threshold)]

# Regroupement des données par année de sortie et calcul de la médiane des revenus pour chaque année
revenues_by_year = filtered_data.groupby(filtered_data['release_date'].dt.year)['revenue'].median() / 1e6

# Courbe pour montrer l'évolution des revenus des films au fil du temps
plt.figure(figsize=(10, 6))
plt.plot(revenues_by_year.index, revenues_by_year, marker='o', linestyle='-')
plt.xlabel('Année de sortie')
plt.ylabel('Revenus médians des films (en millions de dollars)')
plt.xticks(rotation=45)
plt.title('Évolution des Revenus Médians des Films par Année de Sortie (2000-2022) - Sans valeurs aberrantes')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

Cet histogramme affiche l'évolution des revenues suivant les années (2000 à 2020) .
Cet histogramme nous indique que le film qui est a eu le plus de revenus a été produit dans **les années 2003** et celui qui a eu le moins de revenus à été produit dans **les années 2020**.


<font color="orange"><b><h1> 9. Matrice de corrélation entre la moyenne de vote et la popularité </h1></b></font>

```python

selected_columns = data[['vote_average', 'popularity']]

correlation_matrix = selected_columns.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)


plt.title("Matrice de Corrélation entre Vote Average et Popularité")
plt.show()
```

On remarque une très **faible correlation entre la popularité des films et la moyenne des votes**.

Cela signifie que les évaluations moyennes attribuées par les spectateurs ne sont pas fortement liées à la popularité du film.

Ce qui veut dire que le fait qu'un film soit très populaire ne garantit pas nécessairement qu'il obtiendra des évaluations élevées de la part du public.
