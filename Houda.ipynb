<h1>La disponibilité de vélos et de stands pour la station de vélo correspondant à l'index donné</h1>
<a href="#Sommaire" style="display: block; text-align: center; font-weight: bold; color: #a2d2ff;">Accéder au sommaire</a>

```python
def get_station_availability(df, station_index):
    
    # Extraire les informations de la station de vélo correspondant à l'index donné
    station_info = df.loc[station_index]

    # Extraire la disponibilité de vélos et de stands à partir des données de 'totalStands'
    bikes_avail = station_info['totalStands']['availabilities']['bikes']
    stands_avail = station_info['totalStands']['availabilities']['stands']
    
    # Retourner la disponibilité de vélos et de stands sous forme de tuple
    return bikes_avail, stands_avail
# Exemple d'utilisation de la fonction pour obtenir la disponibilité de la station en index 0
bikes_avail, stands_avail = get_station_availability(df, 0)
print("La station de vélo dispose de", bikes_avail, "vélos disponibles et de", stands_avail, "emplacements de stationnement disponibles.")
```

<div>
    <h1>Proportion de vélos mécaniques et électriques disponibles selon son index:</h1>
    <ul>
        <li> Le calcul de la proportion de vélos mécaniques et électriques disponibles dans une station permet d'obtenir une indication sur la diversité de l'offre de vélos en libre-service dans cette station. En effet, une proportion élevée de vélos électriques peut indiquer que la station est bien équipée pour répondre aux besoins des utilisateurs souhaitant utiliser ce type de vélos, tandis qu'une proportion élevée de vélos mécaniques peut indiquer une offre plus axée sur les besoins des utilisateurs souhaitant utiliser des vélos plus traditionnels.
        </li>
        <li> Ces informations peuvent être utiles pour évaluer la qualité de l'offre de vélos en libre-service dans une station et pour orienter les actions d'amélioration ou de développement du réseau. Par exemple, si une station présente une proportion élevée de vélos électriques, il peut être intéressant d'envisager de renforcer l'offre en vélos mécaniques pour offrir une plus grande diversité de choix aux utilisateurs. À l'inverse, si une station présente une proportion élevée de vélos mécaniques, il peut être intéressant de proposer plus de vélos électriques pour répondre aux besoins des utilisateurs souhaitant ce type de vélos.
        </li>
    </ul>
</div>
<a href="#Sommaire" style="display: block; text-align: center; font-weight: bold; color: #a2d2ff;">Accéder au sommaire</a>

```python
def prop_bikes_by_type(station_index):
    # Extraire les informations de la station de vélo en index donné
    station_info = df.loc[station_index]
    
    # Extraire la disponibilité des vélos mécaniques et électriques à partir des données de 'totalStands'
    mech_bikes_avail = station_info['totalStands']['availabilities']['mechanicalBikes']
    elec_bikes_avail = station_info['totalStands']['availabilities']['electricalBikes']
    
    # Calculer la proportion de vélos mécaniques et électriques disponibles
    total_bikes = mech_bikes_avail + elec_bikes_avail
    if total_bikes != 0:
        prop_mech_bikes = mech_bikes_avail / total_bikes
        prop_elec_bikes = elec_bikes_avail / total_bikes
    else:
        prop_mech_bikes = 0
        prop_elec_bikes = 0
        
    # Afficher la proportion de vélos mécaniques et électriques disponibles
    print("Proportion de vélos mécaniques disponibles:", prop_mech_bikes)
    print("Proportion de vélos électriques disponibles:", prop_elec_bikes)
prop_bikes_by_type(5);
```

<div>
    <h1>géolocalisation des stations de vélos en libre-service</h1>
    <p>La géolocalisation d'une station peut permettre de :</p>
    <ul>
        <li>Visualiser la position de la station sur une carte pour aider les utilisateurs à trouver une station à               proximité.
        </li>
        <li>Calculer la distance entre la station et d'autres points d'intérêt, tels que les gares, les arrêts de               bus ou les lieux touristiques, pour faciliter l'utilisation combinée de différents modes de transport.
        </li>
        <li>Évaluer la densité de stations dans une zone géographique donnée et identifier les zones où il pourrait             être nécessaire d'implanter de nouvelles stations.
        </li>
        <li>Optimiser les trajets des équipes de maintenance pour réduire les temps de déplacement et d'intervention             en cas de besoin.
        </li>
    </ul>
<p>En somme, la géolocalisation des stations de vélos en libre-service est un élément clé de la gestion de ces réseaux et permet d'améliorer leur fonctionnement pour les utilisateurs.</p>
    <a href="#Sommaire" style="display: block; text-align: center; font-weight: bold; color: #a2d2ff;">Accéder au sommaire</a>

```python
def get_station_location(df, contractName):
    # Extraire les informations de la station de vélo correspondant à la contractName
    station_info = df.loc[df['contractName'] == contractName].iloc[0]
    
    # Extraire les coordonnées de latitude et de longitude
    latitude = station_info['position']['latitude']
    longitude = station_info['position']['longitude']
    
    # Retourner l'URL pour la localisation de la station de vélo
    return f"https://www.google.com/maps/place/{latitude},{longitude}"

location = get_station_location(df,'lyon')

# Afficher l'URL pour la localisation de la station de vélo
print(location)

```

<div>
    <h1>Station ouverte ou fermée ?</h1>
    <ul>
        <li>Vérifier la disponibilité d'une station ouverte ou fermée peut être important pour les utilisateurs du système de vélos en libre-service, car cela leur permet de savoir si la station est actuellement utilisable ou non. Cela peut être particulièrement utile pour les personnes qui cherchent à louer un vélo à une station spécifique.
        </li>
        <li>De plus, pour les opérateurs du système, savoir si une station est ouverte ou fermée peut être important pour la gestion des stocks de vélos, l'entretien et la planification de la maintenance.
        </li>
    </ul>
</div>
<a href="#Sommaire" style="display: block; text-align: center; font-weight: bold; color: #a2d2ff;">Accéder au sommaire</a>

```python
def station_disponibilite(df, contractName):
    station_info = df.loc[df['contractName']==contractName].iloc[0]
    if station_info['connected'] and not station_info['overflow'] and station_info['status'] == 'OPEN':
        message = "La station de vélo est ouverte."
    else:
        message = "La station de vélo n'est pas disponible pour le moment."
    return message
print(station_disponibilite(df, 'lyon'))
```
